# Quality Assessment of Food Knowledge Graph

```python
# Import required libraries
import rdflib
from rdflib import Graph, Literal, Namespace, URIRef
from pyshacl import validate

shapes_g = Graph()
RDF = Namespace('http://www.w3.org/1999/02/22-rdf-syntax-ns#')
RDFS = Namespace('http://www.w3.org/2000/01/rdf-schema#')
XSD = Namespace('http://www.w3.org/2001/XMLSchema#')
SCHEMA = Namespace('http://schema.org/')
OWL = Namespace('http://www.w3.org/2002/07/owl#')
DBO = Namespace('http://dbpedia.org/ontology/')
DBR = Namespace('http://dbpedia.org/resource/')
SH = Namespace('http://www.w3.org/ns/shacl#')
KGSH = Namespace('http://kg-course.io/food-nutrition/shacl/')

shapes_g.bind('rdf', RDF)
shapes_g.bind('rdfs', RDFS)
shapes_g.bind('xsd', XSD)
shapes_g.bind('schema', SCHEMA)
shapes_g.bind('owl', OWL)
shapes_g.bind('dbo', DBO)
shapes_g.bind('dbr', DBR)
shapes_g.bind('sh', SH)
shapes_g.bind('kgsh', KGSH)

# Define SHACL shapes for quality assessment
# Shape for Recipe class
shapes_g.add((KGSH.RecipeShape, RDF.type, SH.NodeShape))
shapes_g.add((KGSH.RecipeShape, SH.targetClass, SCHEMA.Recipe))

# Define name property constraints for RecipeShape
shapes_g.add((KGSH.RecipeShape, SH.property, KGSH.RecipeShape_name))
shapes_g.add((KGSH.RecipeShape_name, SH.path, SCHEMA.name))
shapes_g.add((KGSH.RecipeShape_name, SH.datatype, XSD.string))
shapes_g.add((KGSH.RecipeShape_name, SH.minCount, Literal(1)))

# Define recipeIngredient property constraints for RecipeShape
shapes_g.add((KGSH.RecipeShape, SH.property, KGSH.RecipeShape_recipeIngredient))
shapes_g.add((KGSH.RecipeShape_recipeIngredient, SH.path, SCHEMA.recipeIngredient))
shapes_g.add((KGSH.RecipeShape_recipeIngredient, SH.datatype, XSD.string))
shapes_g.add((KGSH.RecipeShape_recipeIngredient, SH.minCount, Literal(1)))
```
